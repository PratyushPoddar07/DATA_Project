```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import linear_model
```

```python
df  = pd.read_csv("C:/Users/shuva/Desktop/Ml_learn/homeprices.csv")
df
```

```python
plt.title('Area_vs_Price')
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='red',marker='+')
```

```python
df['bedrooms'] = df['bedrooms'].fillna(df['bedrooms'].mean())
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area','bedrooms','age']],df.price)
```

```python
df.dtypes
```

```python

predicted_price = reg.predict(np.array([[2600, 1, 20]]))  # ✅ Correct usage
predicted_price
```
