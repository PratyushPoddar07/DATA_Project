```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\shuva\Downloads\titanic.csv")
df.head()
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis =1,inplace =True)
df.head()
```

```python
X = df.drop(['Survived'],axis = 1)
X
```

```python
y = df['Survived']
y
```

```python
dummies = pd.get_dummies(X.Sex)
dummies.head()
```

```python
X = pd.concat([X,dummies],axis = 1)
X
```

```python
## I am dropping male columns and sex columns as well because of  dummy variable trap theory.Onc column is enough to represent male vs female
```

```python
X.drop(['Sex','male'],axis =1,inplace =True)
X.head()
```

```python
X.columns[X.isna().any()]
```

```python
X.Age[:10]
```

```python
X.Age = X.Age.fillna(X.Age.mean())
X.head(10)
```

## Train test split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3,random_state =42)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model = GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test,y_test)
```

```python
model.predict_proba(X_test[:10])
```

## Calculate the score using cross validation

```python
from sklearn.model_selection import cross_val_score
```

```python
val = cross_val_score(model,X_train,y_train,cv =5,scoring='accuracy')
```
