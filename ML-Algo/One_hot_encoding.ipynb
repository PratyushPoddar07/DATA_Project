```python
model.score(X,y)
```

```python
# -> now apply dummies variable for one hot encoder
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(drop='first', sparse_output = False)
```

```python
# drop town 
df = df.drop(['town'],axis =1)
df
```

```python
encoded_features =  ohe.fit_transform(df[['town_label']])
```

```python
encoded_df = pd.DataFrame(encoded_features, columns = ohe.get_feature_names_out(['town_label']))
encoded_df
```

```python
# combine with original DataFrame
df_final = pd.concat([df.drop(columns=['town_label']),encoded_df],axis =1)
df_final
```

```python
X = df_final.drop(columns = ['price']).values
X
```
