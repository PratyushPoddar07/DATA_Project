```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("C:/Users/shuva/Desktop/Ml_learn/carprices.csv")
df.head()
```

## Car Mileage vs Call Sell Price

```python
plt.scatter(df['Mileage'],df['Sell Price($)'],color = 'red',label = 'Mileage')
```

## Car age vs Car sell Price

```python
plt.scatter(df['Age(yrs)'],df['Sell Price($)'],color = 'blue',label = 'Age')
```

```python
X = df.drop('Sell Price($)',axis =1)
X
```

```python
y = df['Sell Price($)']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size =0.2,random_state =10)
```

```python
len(X_train)
len(y_train)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
clf = LinearRegression()
```

```python
clf.fit(X_train,y_train)
```

```python
clf.predict(X_test)
```

```python
y_test
```

```python
clf.score(X_test,y_test)
```
