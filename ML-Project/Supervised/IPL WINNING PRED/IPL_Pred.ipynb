## DATA MODIFICATION 

```python
X = ["city","toss_decision","result","dl_applied"]
for i in X :
    print("--------------")
    print(df[i].unique())
    print(df[i].value_counts())
```

```python
df.drop(["id","Season","city","date","player_of_match","umpire1","venue","umpire2"],axis=1,inplace =True)
df.head()
```

## Splitting Dataset

```python
X = df.drop(['winner'],axis =1)
y =  df['winner']
```

```python
X = pd.get_dummies(X, ["team1","team2","toss_winner","toss_decision","result"],drop_first = True)
```

```python
X.head()
```

## LAbel Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
y = le.fit_transform(y)
```

## Train-test-split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state =5)
```

## Model training 

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators = 100,min_samples_split=5,max_features="sqrt")
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay,accuracy_score
```

```python
accuracy = accuracy_score(y_test,y_pred)
accuracy
```

## Confusion matrix

```python
confusion_matrix = confusion_matrix(y_test,y_pred)
```

```python
cm = ConfusionMatrixDisplay(confusion_matrix = confusion_matrix)
cm.plot()
```
