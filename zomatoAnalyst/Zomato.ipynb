```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Zomato data .csv")
```

```python
print(df.head())
```

```python
def handleRate(value):
    value = str(value).split('/')
    value = value[0]
    return float(value)

df['rate'] = df['rate'].apply(handleRate)
print(df.head())
```

```python
df.info()
```

```python
sns.countplot(x=df["listed_in(type)"])
plt.xlabel("Type of Restaurent")
```

```python
grouped_data = df.groupby('listed_in(type)')['votes'].sum()
result=pd.DataFrame({'votes': grouped_data})
plt.plot(result,c="green",marker ="o")
plt.xlabel("Type of Restaurent",c = "red",size = 20)
plt.ylabel("Votes",c = "red",size = 20)
```

```python
max_votes=  df['votes'].max()
rest_with_maxVotes = df.loc[df['votes'] == max_votes,'name']
print("Restaurent with max votes:")
print(rest_with_maxVotes)
```

```python
sns.countplot(x=df['online_order'])
```

```python
plt.hist(df['rate'],bins =5)
plt.title("Rating Distribution")
plt.show()
```

```python
coupled_data = df['approx_cost(for two people)']
sns.countplot(x=coupled_data)
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(x='online_order',y='rate',data = df)
```

```python
pivot_table = df.pivot_table(index = 'listed_in(type)',columns ='online_order', aggfunc= 'size',fill_value = 0)
sns.heatmap(pivot_table,annot = True,cmap = "YlGnBu",fmt ='d')
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed In(Type)")
plt.show()
```
